```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results_rows = pd.read_csv("prog1/results_rows.csv")
results_rows = results_rows.groupby(results_rows.index // 15).mean()

results_cols = pd.read_csv("prog2/results_cols.csv")
results_cols = results_cols.groupby(results_cols.index // 15).mean()

plt.plot(
    [1, 2, 4, 8, 16, 32, 64, 128, 256, 512, 1024],
    results_rows["t2"],
    label="Rows processing",
    marker="o",
)

plt.plot(
    [1, 2, 4, 8, 16, 32, 64, 128, 256, 512, 1024],
    results_cols["t2"],
    label="Columns processing",
    marker="o",
)

plt.legend()
plt.title("Comparison between row and column processing \n (log-log scale)")
plt.xlabel("Threads")
plt.xscale("log")
plt.yscale("log")
plt.ylabel("Time (s)")
plt.tight_layout()
```
